```python
sample_data_dic
```

```python
netflix_data=pd.read_csv('./netflix_data.csv')
netflix_data.head()
```

```python
netflix_data.isnull() .sum()
```

```python
years=netflix_data.release_year.to_list()
durations=netflix_data.duration.to_list()
```

```python
movie_dic={
    'years':years,
    'durations':durations
}
```

```python
movie_df=pd.DataFrame(movie_dic)
movie_df.loc[1:10]
```

```python
import warnings
```

```python
import seaborn as sas
```

```python
fig=plt.figure(figsize=(12,6))
plt.style.use('seaborn-v0_8-darkgrid')
sas.lineplot(x='years',y='durations',data=movie_df)
plt.title('NETFLIX_MOVIE_DURATION')
plt.xlabel('Relieas_Years')
plt.ylabel('Relieas_Durations')
plt.show()
```

```python
fig=plt.figure(figsize=(12,6))
plt.style.use('seaborn-v0_8-whitegrid')
sas.lineplot(x='years',y='durations',data=movie_df)
plt.title('NETFLIX_MOVIE_DURATION')
plt.xlabel('Relieas_Years')
plt.ylabel('Relieas_Durations')
plt.legend('durations,years')
plt.show()
```

```python
fig=plt.figure(figsize=(12,6))
plt.style.use('seaborn-v0_8-ticks')
sas.lineplot(x='years',y='durations',data=movie_df)
plt.title('NETFLIX_MOVIE_DURATION')
plt.xlabel('Relieas_Years')
plt.ylabel('Relieas_Durations')
plt.legend('durations,years')
plt.show()
```

```python
netflix_movie_df=netflix_data[netflix_data['type'] == 'Movie']
```

```python
netflix_movie_col_subset=netflix_movie_df[['title','country','genre','release_year','duration']]
```

```python
netflix_movie_col_subset.head()
```

```python
fig=plt.figure(figsize=(12,8))
plt.scatter(x=years,y=durations,data=movie_df)
plt.title('NETFLIX_MOVIE_DURATION')
plt.xlabel('Relieas_Years')
plt.ylabel('Relieas_Durations')
plt.show
```

```python
shorter_movie= netflix_movie_df[netflix_movie_df['duration'] < 60 ]
shorter_movie.head(7)
```

```python
colour=[]
for raw ,columns in netflix_movie_col_subset.iterrows():
    if columns ['genre'] == 'Comedies':
        colour.append('red')
    elif columns['genre'] == 'stand up':
        colour.append('green')
    elif columns['genre'] == 'Documentery':
         colour.append('blue')
    else:
        colour.append('black')
    
    
colour[:10]
```

```python
netflix_movie_df.columns
```

```python
plt.style.use('fivethirtyeight')
fig=plt.figure(figsize=(12,8))
plt.scatter(x='release_year',y='duration',data= netflix_movie_df , c= colour)
plt.title('Scatter Plot of Years vs Durations')
plt.xlabel('Years')
plt.ylabel('Durations')
plt.show
```
